```python
# Add google gemini AI engine
import google.generativeai as genai
q='describe singapore in 10 words'
gemini_api = readkey('gemini')
genai.configure(api_key=gemini_api)
model= genai.GenerativeModel("gemini-2.0-flash") #gemini-1.5-flash obsoleted

# Use gemini to generate response and push to Telegram
r=model.generate_content(q)
r=requests.get(url+f"/sendMessage?chat_id={chat_id}&text={r}")
r
```
